### **Building Recommendation Model**

```python
#first we need to the book table into pivot table
book_df=book.pivot_table(index='UserId',columns='Title',values='Rating')
```

```python
book_df.head()
```

```python
#filling null values with 0
book_df.fillna(0,inplace=True)
```

```python
book_df.head()
```

Cosine Similarity

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#converting this into dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
book['UserId'].unique()
```

```python
user_sim_df.index=book['UserId'].unique()
user_sim_df.columns=book['UserId'].unique()
```
